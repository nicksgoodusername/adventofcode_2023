```python
import re
```

```python
redstr, grnstr, blustr = [fr"\d+(?={x})" for x in ['red', 'green', 'blue']]
colour_list = [redstr, grnstr, blustr]
redmax, grnmax, blumax = [12, 13, 14]
```

```python
### Part 1

game_sum = 5050
game = 0
redstr, grnstr, blustr = [fr"\d+(?={x})" for x in ['red', 'green', 'blue']]
with open('input_2.txt', 'r') as file:
    for line in file:
        impossible_token = 0
        game += 1
        line = line.strip().replace(' ', '')
        sets = line.split(':')[1].split(';')
        for set in sets:
            reds, greens, blues = [int(re.search(x, set).group()) if re.search(x, set) else 0 for x in colour_list]
            if (reds>redmax or greens>grnmax or blues>blumax):
                impossible_token += 1
        if impossible_token > 0:
            game_sum -= game

print(game_sum)

```

```python
### Part 2
total_power = 0
redstr, grnstr, blustr = [fr"\d+(?={x})" for x in ['red', 'green', 'blue']]
with open('input_2.txt', 'r') as file:
    for line in file:
        cubes_min = [0, 0, 0]
        line = line.strip().replace(' ', '')
        sets = line.split(':')[1].split(';')
        for set in sets:
            cubes_pulled = [int(re.search(x, set).group()) if re.search(x, set) else 0 for x in colour_list]
            for x in range(len(cubes_pulled)):
                if cubes_pulled[x] > cubes_min[x]:
                    cubes_min[x] = cubes_pulled[x]
        print(cubes_min)
        power = 1
        for x in cubes_min:
            power = power * x
        total_power += power

print(total_power)

```
