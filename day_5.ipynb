```python
import numpy as np
import re
from tqdm import tqdm
```

```python
map_match = r"(.+)\smap:"
newline_match = r"^\n"
numbers_match = r"(\d+)"
seeds_match = r"(?:seeds:)? (\d+)"
```

```python
### Part 1

section_count = 0
section_number = 0
line_count = 0
total_values = []
names_list = []
section_values = []
file = "./input_5.txt"
with open(file, "r") as input:
    for line in input:
        line_count += 1
        if line_count < 3:
            continue
        name = re.match(map_match, line)
        if name:
            name = name.group(1)
            names_list.append(name)
            section_count += 1
            section_number += 1
            continue
        if re.match(newline_match, line):
            section_count -= 1
            total_values.append(section_values)
            section_values = []
            continue
        if section_count == 1:
            line = re.findall(numbers_match, line)
            line = [int(x) for x in line]
            section_values.append(line)

with open(file, "r") as input:
    for line in input:
        seeds = re.findall(seeds_match, line)
        if seeds:
            seeds = [int(x) for x in seeds]
            break

locations = []
for seed in seeds:
    current_no = seed
    map_count = 1
    for map in total_values:
        line_count = 1
        for line in map:
            # Set the starts, stops and lengths for this line in the map
            source_start = line[1]
            length = line[2]
            source_stop = line[1] + length
            destination_start = line[0]
            if not ((source_start <= current_no) and (source_stop > current_no)):  # If the input number can't be mapped by this
                line_count += 1
                if (line_count > len(map)):  # If final line then there are no mappings and number is unchanged
                    break
                else:  # Otherwise carry on looking
                    continue
            else:  # Otherwise the input number is mapped by this line
                diff = current_no - source_start
                current_no = destination_start + diff
                break
            
        if map_count == len(total_values):
            location_no = current_no
            locations.append(location_no)
        map_count += 1
        
np.min(locations)
```

```python
test_input = """seeds: 79 14 55 13

seed-to-soil map:
50 98 2
52 50 48

soil-to-fertilizer map:
0 15 37
37 52 2
39 0 15

fertilizer-to-water map:
49 53 8
0 11 42
42 0 7
57 7 4

water-to-light map:
88 18 7
18 25 70

light-to-temperature map:
45 77 23
81 45 19
68 64 13

temperature-to-humidity map:
0 69 1
1 0 69

humidity-to-location map:
60 56 37
56 93 4"""

inputs, *blocks = test_input.split("\n\n")
inputs = list(map(int, inputs.split(":")[1].split()))

for block in blocks:  # For each block/map
    ranges = []  # Equivalent to total_values[i]
    for line in block.splitlines()[1:]:
        ranges.append(list(map(int, line.split())))
    new = []
    while seed_ranges:  # Starting from the back of seed_ranges
        start, end = seed_ranges.pop()
        for destination_start, source_start, length in ranges:  # Ranges for each block/map
            os = max(start, source_start)  # The highest of seed_start and source_start
            oe = min(end, source_start + length)  # The lowest of seed_end and source_end
            if os < oe:  # If there is overlap
                new.append((os - source_start + destination_start, oe - source_start + destination_start))  # distance from ss to os mapped, distance from ss to oe mapped
                if os > start:
                    seed_ranges.append((start, os))  # Add the unmapped remainder back to seed_ranges
                if end > oe:
                    seed_ranges.append((oe, end))
                break
        else:
            new.append((start, end))  # If never overlap found then remain unmapped
    seed_ranges = new
print(min(seed_ranges)[0])
```
