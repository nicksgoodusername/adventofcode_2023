```python
import numpy as np
import re
from tqdm import tqdm
```

```python
map_match = r"(.+)\smap:"
newline_match = r"^\n"
numbers_match = r"(\d+)"
seeds_match = r"(?:seeds:)? (\d+)"
```

```python
### Part 1

section_count = 0
section_number = 0
line_count = 0
total_values = []
names_list = []
section_values = []
file = "./input_5.txt"
with open(file, "r") as input:
    for line in input:
        line_count += 1
        if line_count < 3:
            continue
        name = re.match(map_match, line)
        if name:
            name = name.group(1)
            names_list.append(name)
            section_count += 1
            section_number += 1
            continue
        if re.match(newline_match, line):
            section_count -= 1
            total_values.append(section_values)
            section_values = []
            continue
        if section_count == 1:
            line = re.findall(numbers_match, line)
            line = [int(x) for x in line]
            section_values.append(line)

with open(file, "r") as input:
    for line in input:
        seeds = re.findall(seeds_match, line)
        if seeds:
            seeds = [int(x) for x in seeds]
            break

locations = []
for seed in seeds:
    current_no = seed
    map_count = 1
    for src2seed_map in total_values:
        line_count = 1
        for line in src2seed_map:
            # Set the starts, stops and lengths for this line in the map
            source_start = line[1]
            length = line[2]
            source_stop = line[1] + length
            destination_start = line[0]
            if not ((source_start <= current_no) and (source_stop > current_no)):  # If the input number can't be mapped by this
                line_count += 1
                if (line_count > len(src2seed_map)):  # If final line then there are no mappings and number is unchanged
                    break
                else:  # Otherwise carry on looking
                    continue
            else:  # Otherwise the input number is mapped by this line
                diff = current_no - source_start
                current_no = destination_start + diff
                break
            
        if map_count == len(total_values):
            location_no = current_no
            locations.append(location_no)
        map_count += 1
        
np.min(locations)
```

```python
### Part 2

map_match = r"(.+)\smap:"
newline_match = r"^\n"
numbers_match = r"(\d+)"
seeds_match = r"(?:seeds:)? (\d+)"

section_count = 0
section_number = 0
line_count = 0
total_values = []
names_list = []
section_values = []
file = "./input_5.txt"
with open(file, "r") as input:
    for line in input:
        line_count += 1
        if line_count < 3:
            continue
        name = re.match(map_match, line)
        if name:
            name = name.group(1)
            names_list.append(name)
            section_count += 1
            section_number += 1
            continue
        if re.match(newline_match, line):
            section_count -= 1
            total_values.append(section_values)
            section_values = []
            continue
        if section_count == 1:
            line = re.findall(numbers_match, line)
            line = [int(x) for x in line]
            section_values.append(line)

with open(file, "r") as input:
    for line in input:
        seeds = re.findall(seeds_match, line)
        if seeds:
            seeds = [int(x) for x in seeds]
            break
seed_ranges = [(start, start + sr_length) for start, sr_length in np.array_split(seeds, len(seeds)/2)]  # Remember the ends are non-inclusive!

ranges = seed_ranges  # Start by inputting the seed ranges
for src2seed_map in total_values:  # For each map
    mapped_ranges = []  # Initialise mapped values
    while ranges:
        range_start, range_end = ranges.pop()  # Get the final range start and end
        for destination_start, source_start, map_length in src2seed_map:  # For each line in the map
            ovlp_s = max(range_start, source_start)  # The start of the overlap between the input range and the map's source range
            ovlp_e = min(range_end, source_start + map_length)  # The end of the overlap
            if ovlp_s < ovlp_e:  # If thers is overlap
                mapped_ranges.append((ovlp_s - source_start + destination_start, ovlp_e - source_start + destination_start))  # Map the overlapped values
                if range_start < ovlp_s:  # If there is end overhang put the unmapped values back into ranges
                    ranges.append((range_start, ovlp_s))
                if range_end > ovlp_e:  # Same for beginning overhang
                    ranges.append((ovlp_e, range_end))
                break
        else:  # If no overlap for input ranges for this map
            mapped_ranges.append((range_start, range_end))
    ranges = mapped_ranges

ranges = np.array(ranges)
print(min(ranges[:, 0]))
```
